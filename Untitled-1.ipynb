To apply a transformation, the skewness must not be within [-0.5, 0.5].
In such cases, the distribution is said to be asymmetric —
it has a right tail when the skewness is positive,
and a left tail when the skewness is negative.

Pour effectuer une transformation, il faut que ma skewness ne soit pas comprise entre [-0.5, 0.5].
Dans ces cas, on dit que la distribution est asymétrique :
elle présente une queue à droite lorsque la skewness est positive,
ou une queue à gauche lorsqu’elle est négative.

14)Target encoding 
A)Initialization of the KFold
    Initialisation de KFold



```python
from sklearn.model_selection import KFold

kf=KFold(n_splits=7,shuffle=True, random_state=42)
#A
col_tar=['Exterior','Neighborhood']

for i in col_tar:
    x_train[f"{i}_ta"]=0

for train_id, val_id in kf.split(x_train):
    for col in col_tar:
        means=log_target.iloc[train_id].groupby(x_train.iloc[train_id][col]).mean()
        x_train.loc[val_id,f"{col}_ta"]=x_train.loc[val_id,col].map(means)

#Application sur x_test
for col in col_tar:
    means_global = log_target.groupby(x_train[col]).mean()
    global_mean = log_target.mean()
    df_test[f"{col}_ta"] = df_test[col].map(means_global).fillna(global_mean)

#Remove the columns with big rendancy with certain columns in the model actually 
#Suppression des colonnes avec une grande redondance avec certaines colonnes dans le modèle actuel
x_train.drop(['Exterior1st','Exterior2nd','Neighborhood',"Exterior"],axis=1,inplace=True)
df_test.drop(['Exterior1st','Exterior2nd','Neighborhood',"Exterior"],axis=1,inplace=True)
```

Since my target encoding is based on a portion of the dataset using K-Fold, some categories may appear in one fold but not in another, which produces NaN.
I therefore replace them with 0.
Additionally, categories that exist only in x_test will also produce NaN after encoding, which I also replace with 0 to ensure the model can handle all columns.

Comme mon target encoding se base sur une partie du dataset grâce à K-Fold, certaines catégories peuvent apparaître dans un fold mais pas dans un autre, ce qui génère des NaN.
Je les remplace donc par 0.
De plus, les catégories présentes uniquement dans x_test produiront également des NaN après l’encodage, que je remplace aussi par 0 pour garantir que le modèle puisse gérer toutes les colonnes.
